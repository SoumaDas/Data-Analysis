# Line plot
show relations between two variables(cont. or discrete)

```python
x = np.linspace(1,50)
x
```

```python
y1 = np.array((2*x)+1)
print(y1)
print('\n\n')
y2 = np.array(x**2+10)
print(y2)
print('\n\n')
y3 = np.array(np.log2(x))
print(y3)
print('\n\n')
y4 = np.array(2**x)
print(y4)
print('\n\n')
```

plt.figure() it is a fuction which is callled for figure

```python
plt.plot(x,y1)
plt.title("Linear function :") # it gives a title of the graph
plt.xlabel('X') # lebeling the x axis
plt.ylabel('Y=2X+1') #lebeling the y axis
plt.show()
```

```python
plt.xlim(-50,50) # it it used to give the range of x-axis if range is big then the graph will be zoom out
plt.plot(x,y1)
```

```python
plt.xlim(20,50) # it it used to give the range of x-axis if range is big then the graph will be zoom in
plt.plot(x,y2)
plt.savefig(r'D:\globsyn\images\test.jpg') #save the images
```

```python
plt.figure(figsize=(12,5)) # figsize will change the size of the graph
plt.xticks(np.arange(1,50,2)) 
```

```python
plt.figure(figsize = (12,5))
plt.plot(x,y2, marker = 'H', mfc = 'y', mec = 'c', c = '#20e2ff', ls= ':')
# mfc = marker face colour, mec = marker edge colour, c = colour of the line
```

# sob plotting
plotting different graphs in one graph

```python
plt.subplot(221) #dividing the graph into 2 columns and 2 rows and 1 menas accessing 1st plot
plt.plot(x,y1)
plt.title('Linear Function')
plt.subplot(222) # accessing 2nd plot
plt.plot(x,y2)
plt.title('Quadratic Function')
plt.subplot(223) # accessing 3rd plot
plt.plot(x,y3)
plt.title('Log Function')
plt.subplot(224) #accessing 4th plot
plt.plot(x,y4)
plt.title('Exponenrial Function')
plt.tight_layout()
plt.show()
```

```python
#multiple lines in a single canvas
plt.figure()
plt.plot(x,y1, label = 'Linear')
plt.plot(x,y2, label = 'Quadratic')
plt.plot(x,y3, label = 'Log')
plt.legend() # lebeling the lines
plt.show()
```

# bar graph
it shows the magnitube of some values in a graph

```python
fruit = { 'CNN (95%)': 95 , 'Transfer Learning (83%)': 83, 'LSTM-RNN (90%)': 90}
df = pd.Series(fruit)
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df is your DataFrame with model accuracies
# For demonstration, I'm generating some sample data
# df = pd.Series([90, 85, 78, 92, 88], index=['Model A', 'Model B', 'Model C', 'Model D', 'Model E'])

# Create a colormap instance
cmap = plt.get_cmap('viridis')

# Normalize values to the colormap range
colors = cmap(np.linspace(0, 1, len(df)))

plt.figure(figsize=(10, 6))
plt.title("Bar Chart of Every Model's Accuracy")
plt.xlabel('Model')
plt.ylabel('Accuracy in Percentage')

# Use the 'colors' array for bar color assignment
plt.bar(df.index, df.values, color=colors)

plt.grid(axis='y')
plt.show()
```
